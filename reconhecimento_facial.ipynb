```python
from mtcnn import mtcnn
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
```

```python
img = plt.imread('/content/imagem12.jpeg')
# Convert image to RGB if it's in RGBA format
if img.shape[-1] == 4:
    img = img[..., :3]
detector = mtcnn.MTCNN()
```

```python
faces = detector.detect_faces(img, min_face_size = 20)
```

```python
def imagens (img,faces):
    plt.imshow(img)
    ax = plt.gca()
    for face in faces :
        x,y,width,height = face['box']
        confidence =face['confidence']
        rect = Rectangle((x,y),width,height,color='blue',fill=False,lw=4)
        ax.add_patch(rect)
        ax.text(x,y-10, f"{confidence:.2f}", color = 'red', fontsize=12, weight = 'bold')
    plt.axis('off')
    plt.imshow(img)
```

```python
rosto = imagens(img=img,faces=faces)
```
